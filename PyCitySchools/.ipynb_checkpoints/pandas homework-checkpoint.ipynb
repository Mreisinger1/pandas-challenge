```python
#dependencies and set up
import pandas as pd

#file to load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

#read school and student data file and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

#combine data
school_data_complete = pd.merge(student_data, school_data)
```

District Summary

```python
#total number of schools
school_count = school_data["school_name"].count()
school_count
```

```python
#total number of students
student_count = school_data_complete["Student ID"].count()
student_count
```

```python
#total budget
total_budget = school_data["budget"].sum()
total_budget
```

```python
#average math score
average_mathscore = school_data_complete["math_score"].mean()
average_mathscore
```

```python
#average reading score
average_readingscore = school_data_complete["reading_score"].mean()
average_readingscore
```

```python
#calculate the percentage of students with a passing math score 
passing_mathscore = len(school_data_complete[school_data_complete["math_score"] >= 70])
passing_mathscore
```

```python
#calculate the percentage of students with a passing reading score
passing_readingscore = len(school_data_complete[school_data_complete["reading_score"] >= 70])
passing_readingscore
```

```python
#calculate the number of students who passed both math and reading
passing_mathreading = (passing_mathscore + passing_readingscore)
passing_mathreading
```

```python
#calculate the passing math percentage
students_passing_math = school_data_complete.loc[school_data_complete["math_score"] >= 70]
number_students_passing_math = students_passing_math["Student ID"].count()
percent_passing_math = (number_students_passing_math / student_count) * 100
percent_passing_math
```

```python
#calculate the passing reading percentage
students_passing_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
number_students_passing_reading = students_passing_reading["Student ID"].count()
percent_passing_reading = (number_students_passing_reading / student_count) * 100
percent_passing_reading
```

```python
#overall passing percentage
overall_passing = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]["Student ID"].count()/student_count*100
overall_passing
```

```python
#create a dataframe from above data
district_summary = pd.DataFrame({ "Total Schools": [school_count],
    "Total Students": [student_count],
    "Total Budget": [total_budget],
    "Average Math Score": [average_mathscore],
    "Average Reading Score": [average_readingscore],
    "Passing Math": [percent_passing_math],
    "Passing Reading": [percent_passing_reading],
    "Overall Passing": [overall_passing]})
```

```python
district_summary[["Total Schools", "Total Students","Total Budget","Average Math Score",
                     "Average Reading Score","Passing Math","Passing Reading", "Overall Passing"]]
```

School Summary

```python
#school name
school_name = school_data_complete.set_index("school_name").groupby(["school_name"])
```

```python
#school type
school_types = school_data.set_index(["school_name"])["type"]
school_types
```

```python
#total students per school 
total_students = school_data.set_index(["school_name"])["size"]
total_students
```

```python
#total school budget 
total_budget = school_data.set_index(["school_name"])["budget"]
total_budget
```

```python
#budget per student 
student_budget = total_budget / total_students
student_budget
```

```python
#average math scores per school 
average_mathscore_perschool = school_data_complete.groupby(["school_name"]).mean()["math_score"]
average_mathscore_perschool
```

```python
#calculate the passing math percentage
passing_math_percentage = school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name")["Student ID"].count()
passing_math_percent = passing_math_percentage / total_students *100
passing_math_percent
```

```python
#average reading score per school
average_readingscore_perschool = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
average_readingscore_perschool
```

```python
#calculate the passing reading percentage
passing_read_percentage = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name")["Student ID"].count()
passing_read_percent = passing_read_percentage / total_students *100
passing_read_percent
```

```python
#percentage that passed both math and reading 
passing_mathreading_percent = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)].groupby("school_name")["Student ID"].count()
passing_mathreading_percentage = passing_mathreading_percent / total_students *100
passing_mathreading_percentage
```

```python
#create a dataframe to hold the above results
school_summary = pd.DataFrame({ "School Type": school_types, "Total School Budget": total_budget,
                               "Budget Per Student": student_budget,
    "Average Math Score": average_mathscore_perschool, "Average Reading Score": average_readingscore_perschool,
    "Passing Math": passing_math_percent,"Passing Reading": passing_read_percent,
    "Overall Passing": passing_mathreading_percentage})
```

```python
school_summary[["School Type","Total School Budget", "Budget Per Student","Average Math Score","Average Reading Score","Passing Math","Passing Reading","Overall Passing"]]

```

Highest Performing Schools

```python
#highest performing schools 
highest_performing = school_summary.sort_values(["Overall Passing"], ascending = False)
highest_performing
```

Lowest Performing Schools

```python
#lowest performing schools 
lowest_performing = school_summary.sort_values(["Overall Passing"], ascending = True)
lowest_performing
```

Scores by School Spending 

```python
#create bins to group school spending
spending_bins = [0, 565, 600, 635, 675]
group_names = ["<$565", "$565-600", "$600-635", "$635-675"]
```

```python
school_data_complete["spending_bins"] = pd.cut(school_data_complete["budget"]/school_data_complete["size"], spending_bins, labels = group_names)

by_spending = school_data_complete.groupby("spending_bins")

avg_math = by_spending["math_score"].mean()
avg_read = by_spending["reading_score"].mean()
pass_math = school_data_complete[school_data_complete["math_score"] >= 70].groupby("spending_bins")["Student ID"].count()/by_spending["Student ID"].count()*100
pass_read = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("spending_bins")["Student ID"].count()/by_spending["Student ID"].count()*100
overall = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)].groupby("spending_bins")["Student ID"].count()/by_spending["Student ID"].count()*100
               
scores_by_spend = pd.DataFrame({ "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    "Passing Math": pass_math,
    "Passing Reading": pass_read,
    "Overall Passing": overall })
```

```python
#create data frame from above data
scores_by_spend[[ "Average Math Score","Average Reading Score", "Passing Math","Passing Reading", "Overall Passing"]]
```
